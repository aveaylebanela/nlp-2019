```python
data = corpus[0:17]
regular_parse = []
for line in data:
    regex = '[.?!]\s[А-ЯЁ]'
    sents = re.sub('([.?!])\s([А-ЯЁ])', r'\1±#к0$161лi$§\2', line)
    sents = re.split(r'±#к0$161лi$§', sents)
    print(sents)
    for sent in sents:
        regular_parse.append(sent)
#плохо что мы не сплитим по скобкам
```

**Попробуем тоже-мне-метрику**

```python
print('Manual pase - ', len(manual_parse))
print('Regular parse - ', len(regular_parse))
len(set(manual_parse) & set(regular_parse)) / len(set(manual_parse) | set(regular_parse))
```

**Попробуем метрику с ф-мерой**

```python
tp = 0
fp = 0
fn = 0

for sent in regular_parse:
    if len(sent_tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(regular_parse)-1):
    sent1, sent2 = regular_parse[i], regular_parse[i+1]
    sent = ' '.join([sent1, sent2])
    if len(sent_tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```
